```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
```

```python
from keras.layers import Dropout
```

```python
train_dates = pd.to_datetime(datos['FECHA'])
```

```python
cols = list(datos)[1:12]
```

```python
df_for_training = datos[cols].astype(float)
```

```python
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
df_for_training_scaled = scaler.transform(df_for_training)
```

```python
len(df_for_training_scaled)
```

```python
trainX = []
trainY = []
```

```python
n_future = 1
n_past = 12
```

```python
for i in range(n_past, len(df_for_training_scaled) - n_future +1):
    trainX.append(df_for_training_scaled[i - n_past:i, 0:df_for_training.shape[1]])
    trainY.append(df_for_training_scaled[i + n_future -1:i + n_future, 0])
```

```python
trainX, trainY = np.array(trainX), np.array(trainY)
```

```python
print('trainX shape == {}.'.format(trainX.shape))
print('trainY shape == {}.'.format(trainY.shape))
```

```python
model = Sequential()
model.add(LSTM(128, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
#model.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
#model.add(LSTM(32, activation='relu', return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(trainY.shape[1]))

model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
history = model.fit(trainX, trainY, epochs=30, batch_size=120, validation_split=0.1, verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
```
